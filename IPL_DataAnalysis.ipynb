# $Batting Analysis$

**Q1: Which batsman has the highest strike rate (min 500 balls)?**

```python
df.columns
```

```python
df['batter'].value_counts()
```

```python
df['batsman_runs']
```

```python
runs = df.groupby('batter')['batsman_runs'].sum()
```

```python
# ball faced 
balls = df.groupby('batter').size()
```

```python
batsman_stats = pd.DataFrame({
    'runs':runs,
    'balls':balls
})
```

```python
batsman_stats.head(10)
```

```python
batsman_stats = batsman_stats[batsman_stats['balls'] >= 500]
```

```python
batsman_stats['strike_rate'] = (batsman_stats['runs']/batsman_stats['balls'])*100
```

```python
top_sr = batsman_stats.sort_values(by='strike_rate', ascending=False)
```

```python
top10 = top_sr.head(10)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=top10.index,
    y=top10['strike_rate'],
    palette="viridis"
)

plt.xticks(rotation=45, ha='right')

plt.title("Top 10 Batsmen by Strike Rate (Min 500 Balls)", fontsize=16, fontweight='bold')
plt.xlabel('Batsman')
plt.ylabel('Strike Rate')


for i, v in enumerate(top10['strike_rate']):
    plt.text(i, v + 1, f"{v:.1f}", ha='center')

# plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x = top10['strike_rate'])
```

**Q2: Who are the most consistent batsmen?**

```python
df.columns
```

```python
df['batter'].value_counts()
```

```python
df['inning'].value_counts()
```

```python
df['player_dismissed'].value_counts()
```

```python
df_out = df[df['player_dismissed'] !=  'Not Out']
#total ball faced
balls = df.groupby('batter').size()
dismissals = df_out['player_dismissed'].value_counts()

consistency = pd.DataFrame({
    'balls':balls,
    'dismissals':dismissals
}).fillna(0)
```

```python
consistency
```

```python
consistency = consistency[consistency['dismissals'] > 0]

consistency['balls_per_dismissal'] = consistency['balls'] / consistency['dismissals']

consistency = consistency[consistency['balls'] >= 500]
top_consistent = consistency.sort_values(by='balls_per_dismissal', ascending=False)
```

```python
top_consistent.head(10)
```

```python
top10 = top_consistent.head(10)
plt.figure(figsize=(12,6))
sns.barplot(
    x=top10.index,
    y=top10['balls_per_dismissal'],
    palette="viridis"
)

plt.title("Most Consistent Batsmen (Min 500 Balls)")
plt.xlabel("Batsman")
plt.ylabel("Balls per Dismissal")

for i, v in enumerate(top10['balls_per_dismissal']):
    plt.text(i, v +0.1, f"{v:.1f}", ha='center')
plt.xticks(rotation = 45)
plt.show()
```

# Bowling Analysis

**Q1: Which bowler has the best economy rate?**

```python
df.columns
```

```python
df['bowler'].value_counts()
```

```python
df['over'].value_counts()
```

```python
runs = df.groupby('bowler')['total_runs'].sum()
```

```python
balls = df.groupby('bowler').size()
```

```python
over = balls/6
```

```python
economy = runs/over
```

```python
bowler_stats = pd.DataFrame({'runs': runs,
                       'balls':balls,
                            'over': over,
                            'economy': economy})
```

```python
bowler_stats
```

```python
bowler_stats = bowler_stats[bowler_stats['balls']>=300]
```

```python
best_economy = bowler_stats.sort_values(by='economy').head(10)
```

```python
best_economy
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    x=best_economy.index,
    y=best_economy['economy'],
    palette='coolwarm'
)
for i, v in enumerate(best_economy['economy']):
    plt.text(i, v +0.1, f"{v:.1f}", ha='center')
plt.title('Best Economy bowler(minimum 300 balls)')
plt.xlabel('Bowler')
plt.ylabel('Economy')
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.scatterplot(
    data=bowler_stats,
    x='balls',
    y='economy'
)

plt.title("Economy vs Balls Bowled")
plt.xlabel("Balls")
plt.ylabel("Economy")
```

**Q3.Which bowler takes wickets most frequently?**


```python
wicket = df.groupby('bowler')['is_wicket'].sum()

```

```python
balls = df.groupby('bowler').size()
```

```python
bowler_stats = pd.DataFrame({'wicket':wicket,
                            'balls': balls})
```

```python
bowler_stats.head(10)
```

```python
bowler_stats = bowler_stats[bowler_stats['wicket']>0]
```

```python
bowler_stats = bowler_stats[bowler_stats['balls']>=300]
```

```python
bowler_stats['strike_rate'] = bowler_stats['balls']/bowler_stats['wicket']
```

```python
best_freq = bowler_stats.sort_values(by='strike_rate')
```

```python
top_bowler = best_freq.head(10)
top_bowler
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x = top_bowler.index , y = top_bowler['strike_rate'] , palette='magma')
plt.xticks(rotation = 45)
plt.title('Highest Wicket in the IPL History')
plt.xlabel('Bowler')
plt.ylabel('Wicket')
for i, v in enumerate(top_bowler['strike_rate']):
    plt.text(i, v +0.1, f"{v:.1f}", ha='center')
plt.show()
```
